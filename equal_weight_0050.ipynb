```python
import numpy as np
import pandas as pd
import requests
import xlsxwriter
import math
import quantstats as qs
import yfinance as yf
```

### Importing Our List of Stocks

```python
stocks = pd.read_csv("0050_stocks.csv")
stocks.head()
```

```python
my_columns = ['Ticker', 'Name', 'Stock Price', 'Market Capitaliztion', 'Number of shares to buy']
rows_list = []
for index, row in stocks.iterrows():
    symbol = row['股票代碼']  # 访问股票代碼列
    name = row['股票名稱']    # 访问股票名稱列
    data = yf.Ticker(f"{symbol}.TW")  # 使用股票代碼构建 Ticker
    stock_data = data.info  # 获取股票的信息
    
    new_list = [
        symbol,
        name,
        stock_data.get('previousClose', None),  
        stock_data.get('marketCap', None),  
        'N/A'
    ]
    rows_list.append(new_list)

final_dataframe = pd.DataFrame(rows_list, columns=my_columns)
final_dataframe
```

```python
portfoil_size = input('Enter the value of your portfolio : ')

try:
    val = float(portfoil_size)
    print(val)
except ValueError:
    print("That's not a number! \nPlease try again:")
    portfoil_size = input('Enter the value of your portfolio : ')
    val = float(portfoil_size)
```

```python
portfoil_size = val/len(final_dataframe.index)
for i in range(len(final_dataframe.index)):
    final_dataframe.loc[i, 'Number of shares to buy'] = math.floor(portfoil_size/final_dataframe.loc[i, 'Stock Price'])
    
final_dataframe
```

```python
writer = pd.ExcelWriter('equal_weight_portfolio.xlsx', engine = "xlsxwriter")
final_dataframe.to_excel(writer, sheet_name = "Equal Weight Portfolio", index = False)
writer.close()
```
